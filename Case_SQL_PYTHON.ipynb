```python
import pandas as pd
from pandasql import sqldf
```

```python
df_casos_covid = pd.read_csv("/content/caso_full.csv")
```

```python
df_casos_covid.shape
```

# 1. Selecione todas as cidades presentes na base de dados;


```python
query = '''
SELECT DISTINCT city
FROM df_casos_covid
WHERE city IS NOT NULL;
'''
```

```python
resultado = sqldf(query)
```

```python
print(resultado)
```

# 2. Encontre a data mais recente em que cada cidade registrou novos casos confirmados

```python
query2 = """
SELECT city, MAX(last_available_date) as data_mais_recente
FROM df_casos_covid
WHERE last_available_confirmed > 0
GROUP BY city;
"""
```

```python
resultado2 = sqldf(query2)
```

```python
print(resultado2)
```

# 3. Encontre a média de mortalidade por estado;

```python
query3 = """
SELECT
    state,
    AVG(last_available_death_rate) as average_mortality_rate
FROM
    df_casos_covid
GROUP BY
    state
"""
```

```python
resultado3 = sqldf(query3)
```

```python
print(resultado3)
```

# 4. Para cada estado, encontre a cidade com a maior taxa de mortalidade

```python
query4 = """
SELECT
    state,
    city,
    MAX(last_available_death_rate) as max_mortality_rate
FROM
    df_casos_covid
GROUP BY
    state
ORDER BY
    state
"""
```

```python
resultado4 = sqldf(query4)
```

```python
print(resultado4)
```

# 5. Obtenha a soma total de novos casos confirmados para cada estado em uma semana epidemiológica específica


```python
query5 ="""
SELECT
    state,
    SUM(new_confirmed) as total_new_confirmed
FROM
    df_casos_covid
WHERE
    epidemiological_week = 202012
GROUP BY
    state
"""
```

```python
resultado5 = sqldf(query5)
```

```python
print(resultado5)
```
